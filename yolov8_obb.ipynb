# Yolo Obb Train

```python
from ultralytics import YOLO

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model_nano_checkp = YOLO('yolov8n-obb.pt')
```

```python
dataset_path = '/content/drive/MyDrive/hack_a_ton/knijki_yolo/dataset.yaml'

results = model_nano_checkp.train(data=dataset_path, epochs=500, patience=20, imgsz=1024, batch=16)
```

```python
metrics = model_nano_checkp.val(data=dataset_path)
for k, v in metrics.results_dict.items():
  print(f'{k}: {v}')
```

# Yolo Inference

```python
images_path = Path('/content/drive/MyDrive/hack_a_ton/test_images')

for i, image in enumerate(list(images_path.glob('*.*'))):

    results = model_nano_checkp(str(image))

    for r in results:
        im_array = r.plot() # plot a BGR numpy array of predictions
        im = Image.fromarray(im_array[..., ::-1]) # RGB PIL image
        plt.figure(figsize=(10, 12))
        plt.imshow(im)
```
